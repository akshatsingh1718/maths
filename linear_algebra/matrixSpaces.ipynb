```python
import numpy as np
import matplotlib.pyplot as plt
```


---
# Column space of a matrix
---

```python
def plot_v(v):
    # matrix S
    S = np.array( [ [3,0],
                    [5,2],
                    [1,2] ] )

    fig = plt.figure()
    ax = fig.gca(projection = '3d')
    xx, yy = np.meshgrid(np.linspace(-5, 5, 5), np.linspace(-5, 5, 5))

    cp = np.cross(S[:, 0], S[:, 1])
    z1 = (-cp[0]*xx - cp[1]*yy)/cp[2]
    ax.plot_surface(xx, yy, z1, alpha=.2)

    ## plot the two vectors from matrix S
    ax.plot([0, S[0,0]], [0, S[1,0]], [0, S[2,0]], 'k')
    ax.plot([0, S[0,1]], [0, S[1,1]], [0, S[2,1]], 'k')

    # plotting vector v
    ax.plot([0, v[0]], [0, v[1]], [0, v[2]], 'r')

    ax.view_init(elev=150, azim=0)
    plt.show()
```

```python
v1 = np.array([[-3], [1], [5]])
v2 = np.array([[1], [7], [3]])
```

```python
# Vector v1 is spanned by S
plot_v(v1)
```

```python
# Vector v2 is not spanned by S
plot_v(v2)
```
