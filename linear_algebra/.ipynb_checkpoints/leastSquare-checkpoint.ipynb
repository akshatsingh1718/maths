# Least-squares via row-reduction
---


```python
m = 10
n = 3

# create data
X = np.random.randn(m,n) # "design matrix"
y = np.random.randn(m,1) # "outcome measures (data)"

np.shape(y)
```

```python
# try directly applying RREF
Xy = Matrix( np.concatenate([X,y],axis=1) )
print( Xy.rref() )
# we just get identity matrix
```

```python
XtX = X.T @ X
Xty = X.T @ y
normEQ = Matrix(np.concatenate([XtX, Xty], axis=1))
Xsol = normEQ.rref()[0]
beta = Xsol[:,-1]
print(np.array(Xsol)), print(' ')
print(beta), print(' ')

# compare to left-inverse
beta2 = np.linalg.inv(X.T @ X) @ Xty
print(beta2), print(' ')

# and with the python solver
beta3 = np.linalg.solve(XtX, Xty)
print(beta3)
```

# Least-squares example 1
---


```python
# data
data = np.array([[-4,0,-3,1,2,8,5,8]]).T
N    = len(data)

# design matrix
X = np.ones([N, 1])

# fit the model
b = np.linalg.solve(X.T@X, X.T@data)
print(f'b : {b}')

# Comparing against the mean
m = np.mean(data)
print(f'mean(data) : {m}')
```

```python
# Computing the model predicted values
yhat = X @ b
print(f'yhat (predicted):{yhat.T}')
print(f'data (actual)   :{data.T}')
# plotting the data and model predicting
plt.plot(np.arange(1, N+1), data,'bs--', label='data')
plt.plot(np.arange(1, N+1), yhat, 'ro--', label='Model predicted')

plt.legend()
plt.show()
```

```python
# new desing matrix
X = np.array([np.arange(0, N)]).T

# fit the model
b = np.linalg.solve(X.T @ X, X.T @ data)

# Model predicting values
yhat = X @ b
print(f'yhat (predicted):{np.round(yhat.T, 4)}')
print(f'data (actual)   :{data.T}')
plt.plot(np.arange(1, N+1), data,'bs--', label='data')
plt.plot(np.arange(1, N+1), yhat, 'ro--', label='Model preidcted')

plt.legend()
plt.show()
```

```python
# new design matrix
X = np.concatenate([np.ones([N, 1]), np.array([np.arange(0, N)]).T], axis=1)

# fit the model
b = np.linalg.solve(X.T @ X, X.T @ data)

# Model prediction
yhat = X @ b

print(f'yhat (predicted):{np.round(yhat.T, 4)}')
print(f'data (actual)   :{data.T}')

# plotting model prediction and actual data
plt.plot(np.arange(1, N+1), data, 'bs--', label='data')
plt.plot(np.arange(1, N+1), yhat, 'ro--', label='Model predicted')

plt.legend()
plt.show()
```

```python
## now with nonlinearity in the design matrix

# design matrix
X = np.concatenate([np.ones([N, 1]), np.array([np.arange(0, N)**2]).T], axis=1)

# fitting the model
b = np.linalg.solve(X.T @ X, X.T @ data)

# predicting
yhat = X @ b

print(f'yhat (predicted):{np.round(yhat.T, 4)}')
print(f'data (actual)   :{data.T}')
# plotting
plt.plot(np.arange(1, N+1), data, 'bs--', label='data')
plt.plot(np.arange(1, N+1), yhat, 'ro--', label='Model predicted')

plt.legend()
plt.show()
```

# Least-squares example 2
---

```python
# load data
data = sio.loadmat('linalg_vectors/leastsquares/EEG_RT_data.mat')
rts = data['rts']# Reaction times
rts = rts[0] 
EEGdata = data['EEGdata'] # EEG Energy
frex = data['frex'] # Frequencies
frex = frex[0]

nTrials = len(rts)
nFrex = len(frex)

# plotting rts
plt.plot(rts, 'ks-')
plt.xlabel('N trials')
plt.ylabel('Response time (ms)')
plt.grid()
plt.show()

# plotting EEG data
plt.imshow(EEGdata, origin='lower')
plt.xlabel('Trial'), plt.ylabel('Frequency')
# low frequencies (close to zero) means slower oscillations and viceversa
# and the yellow colors corespond energies at each of the frequencies.
plt.show()
```

```python
# create and fit the model for one frequency (10 hz)

X = np.concatenate([np.ones([nTrials-1, 1]),                    # intercept
                    np.reshape(rts[0:-1], (nTrials-1, 1)),      # response time
                    np.reshape(EEGdata[10,:-1], (nTrials-1, 1)) # brain 
                   ], axis=1)
# we are taking rts[0,:-1] as we prev reaction time for current
b1 = np.linalg.solve(X.T @ X, X.T @ rts[1:])
b2 = np.linalg.lstsq(X, rts[1:], rcond=None)[0]
print(b1)
print(b2)
```

```python
# fitting model for every frequency
b = np.zeros(len(frex))
t = np.zeros((len(frex), 2))
for f in range(len(frex)):
    X = np.concatenate([np.ones([nTrials,1]),
                        np.reshape(EEGdata[f,:], (nTrials, 1))
                       ], axis=1)
    t[f] = np.linalg.lstsq(X, rts, rcond=None)[0]
    # we are taking only second element bcuz we only want
    # slope which is the releationship b/w EEG enegy and RT
    # first element is only intecept
    b[f] = t[f][1]
    
# plots
plt.figure(figsize=(12,10))
plt.subplot(211)
plt.grid()
plt.plot(frex,b,'rs-')
plt.xlabel('Frequency (Hz)')
plt.ylabel('beta-coefficient')


plt.subplot(223)
plt.plot(EEGdata[8,:],rts,'ks')
# plotting slope
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
intercept = t[8][0]
slope = t[8][1]
y_vals = slope * x_vals + intercept
plt.plot(x_vals, y_vals, 'r-')
plt.ylabel('RTs')
plt.xlabel('Brain energy at ' + str(int(frex[8])) + ' Hz')

plt.subplot(224)
plt.plot(EEGdata[23,:],rts,'ks')
# plotting slope
x_vals = np.array(axes.get_xlim())
intercept = t[23][0]
slope = t[23][1]
y_vals =  slope * x_vals + intercept
plt.plot(x_vals, y_vals, 'r-')
plt.ylabel('RTs')
plt.xlabel('Brain energy at ' + str(int(frex[23])) + ' Hz')
plt.show()
# At 20 hz clicking is fast and at 8 clicking is slow
```
