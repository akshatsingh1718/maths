```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl

mpl.rcParams['text.color'] = 'red'
mpl.rcParams['xtick.color'] = 'r'
mpl.rcParams['ytick.color'] = 'r'
mpl.rcParams['axes.labelcolor'] = 'red'
```


---
# Finding eigenvalues
---


```python
# matrix
A = [ [1,5], [2,4] ]

# eigenvalues
eigenvals = np.linalg.eig(A)
print(f'eigenvalues : {eigenvals[0]}')


# specify two vectors
v1 = np.array([ 1,1 ])    # is an eigenvector!
v2 = np.random.randn(2,1) # unlikely to be an eigenvector
v2 = v2/np.linalg.norm(v2)# unit length for convenience

Av1 = A @ v1
Av2 = A @ v2

# plotting all the vectors
plt.plot([0, Av1[0]], [0, Av1[1]], 'r--', label='Av1')
plt.plot([0, v1[0]], [0, v1[1]], 'r', label='v1')
plt.plot([0, Av2[0]], [0, Av2[1]], 'k--', label='Av2')
plt.plot([0, v2[0]], [0, v2[1]], 'k', label='v2')

plt.axis([-8,8,-8,8])
plt.grid()
plt.legend()
plt.show()
```

#### Code challange: eigenvalues of diagonal and triangular matrices
---

```python
# Eigenvalues for 2x2 Diagonal matrix
D = np.diag(np.array([1, 21]))
eigvals = np.linalg.eig(D)
print(D)
print(' ')
print(f'eigenvals of Diagonal matrix : {eigvals[0]}')
```

```python
# Eigenvalues for 3x3 Diagonal matrix
D = np.diag(np.arange(1,4))
eigvals = np.linalg.eig(D)
print(D)
print(' ')
print(f'eigenvals of Diagonal matrix : {eigvals[0]}')
```

```python
# Eigenvalues for mxm Diagonal matrix
m = 12
D = np.diag(np.arange(m))
eigvals = np.linalg.eig(D)
print(D)
print(' ')
print(f'eigenvals of Diagonal matrix : {eigvals[0]}')
```

```python
# Eigenvalues for mxm upper triangular matrix
m = 4
T = np.triu(np.random.randn(m, m))
eigvals = np.linalg.eig(T)
print(T)
print(' ')
print(f'eigenvals of Diagonal matrix : {eigvals[0]}')
```

```python
# Eigenvalues for mxm lower triangular matrix
m = 4
T = np.tril(np.random.randn(m, m))
eigvals = np.linalg.eig(T)
print(T)
print(' ')
print(f'eigenvals of Diagonal matrix : {eigvals[0]}')
```

#### Code challange: eigenvalues of random matrices
---

```python
m = 40
nIters = 200
for i in range(nIters):
    A = np.random.randn(m, m)
    eigvals = np.linalg.eig(A)[0]
    plt.plot(np.real(eigvals), np.imag(eigvals), 's', markersize=1)
    
plt.axis('square')
plt.grid()
plt.show()
# Here the line in the middle is the real values and img parts are not on zero
# and all the eigenvalues plotted have their mirrored points plottes
```

```python
m = 40
nIters = 200
for i in range(nIters):
    A = np.random.randn(m, m) * 100
    eigvals = np.linalg.eig(A)[0]
    plt.plot(np.real(eigvals), np.imag(eigvals), 's', markersize=1)
    
plt.axis('square')
plt.grid()
plt.show()
```

```python
m = 40
nIters = 200
for i in range(nIters):
    A = np.random.randn(m, m)/100
    eigvals = np.linalg.eig(A)[0]
    plt.plot(np.real(eigvals), np.imag(eigvals), 's', markersize=1)
    
plt.axis('square')
plt.grid()
plt.show()
```

```python
# when dividing by sqrt(m) then all the points lie between 1 to -1
m = 40
nIters = 200
for i in range(nIters):
    A = np.random.randn(m, m)/np.sqrt(m)
    eigvals = np.linalg.eig(A)[0]
    plt.plot(np.real(eigvals), np.imag(eigvals), 's', markersize=1)
    
plt.axis('square')
plt.grid()
plt.show()
```
