# Vector-scalar multiplication
---

```python

v1 = np.array([3, -1])
l = [-.3, 1, 1.5]

def mat_mul_visual(v1, vlm):
    plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')
    plt.plot([0, v1m[0]], [0, v1m[1]], 'r:', label='v1m')
    plt.axis('square')
    axlim = max([max(abs(v1)), max(abs(v1m))])*1.5
    plt.axis((-axlim, axlim, -axlim, axlim))
    plt.grid()
    plt.legend()
    plt.show()
    
    
for i in l:
    v1m = v1 * i
    mat_mul_visual(v1, v1m)
```

# The dot product from a geometric perspective
---

```python
# two vectors
v1 = np.array([ 2,  4, -3 ])
v2 = np.array([ 0, -3, -3 ])

ang = np.arccos(np.dot(v1, v2)/(np.linalg.norm(v1) * np.linalg.norm(v2) ))
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')
plt.axis((-6, 6, -6, 6))
plt.title('angle between vectors {}'.format(ang))
plt.show()
```

```python
alg_dot = np.dot(v1, v2)

geo_dot = np.linalg.norm(v1) * np.linalg.norm(v2) * np.cos(ang)

print(geo_dot, alg_dot)
```

```python
###
# Dot product is how much of 2 vectors are going in the same direction
###
# Generate two vectors 
v1 = np.array([-3, 4, 6])
v2 = np.array([3, 6, -3])

# Generate two scalers
s1 = 2
s2 = 5

# Compute dot product between two vectors 
print(np.dot(v2, np.transpose(v1)))

# Compute dot product between scaled vectors
print(np.dot(v2*s2, np.transpose(v1*s1)))

###
# Here we can see that [ scaled version dot product == original dot product x s1 x s2 ]
###
```

```python
a = np.array([1, 2, 4])
b = np.array([10, 100, 1000])

op1 = np.outer(a, b)

op2 = np.zeros((len(a), len(b)))
for i, i_ele in enumerate(a):
    for j, j_ele in enumerate(b):
        op2[i, j] = i_ele * j_ele
        
print(op1)
print(' ')
print(op2)
```

# Vector cross product
---

```python
# create vectors
v1  = [ -3,  2, 5 ]
v2  = [  4, -3, 0 ]

v3a = np.cross(v1, v2)

v3b = [
    [v1[1]*v2[2] - v1[2]*v2[1]],
    [v1[2]*v2[0] - v1[0]*v2[2]],
    [v1[0]*v2[1] - v1[1]*v2[0]]
]

print('v3a = ', v3a,' v3b = ', v3b)


fig = plt.figure()
ax = fig.gca(projection='3d')

# draw plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
z1 = (-v3a[0]*xx - v3a[1]*yy) / (v3a[2])
ax.plot_surface(xx, yy, z1, alpha=0.2)

# Plot the two vectors
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'b')
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], 'r')

ax.view_init(azim=150,elev=45)

plt.show()
```

```python
def mesh_grid(x, y):
    '''
    numpy meshgrid implementation
    '''
    xx = np.zeros((len(x), len(y)))
    yy = np.zeros((len(x), len(y)))
    
    for i, i_ele in enumerate(x):
        for j, j_ele in enumerate(y):
            xx[j, i], yy[j, i] = i_ele, j_ele
            
    return (xx, yy)

x = [1, 2, 3, 4, 5]
y = [5, 6 ,7 ,8, 9]

xx, yy = np.meshgrid(x, y)
xx1, yy1 = mesh_grid(x, y)

plt.plot(xx, yy, marker='.', color='k', linestyle='none')
plt.show()
plt.plot(xx1, yy1, marker='.', color='r', linestyle='none')
plt.show()
```

# Hermitian transpose (a.k.a. conjugate transpose)
---

```python
# create a complex number
z = np.complex(3, 4)

print(f'Length of z: {np.linalg.norm(z)}')

print(f'Transpose of z : {np.transpose(z)}')

print(f'Hermitian transpose of v : {np.transpose(np.conjugate(z))}')

# complex vector
v = np.array( [ 3, 4j, 5+2j, np.complex(2,-5) ] )
print(v.T)
print(np.transpose(v))
print(np.transpose(np.conjugate(v)))
```

## Code Challange :Dot product with unit vectors
---

```python
# Create 2 random integers vectors (R4)
v1 = np.round(20 * np.random.randn(4))
v2 = np.round(20 * np.random.randn(4))

# Compute the length of individual vectors, and their dot product
v1l = np.sqrt(np.dot(v1, v1))
v2l = np.linalg.norm(v2)
v1v2d = np.abs(np.dot(v1, v2))
print(v1l, v2l, v1v2d)

# 'Normalize' the vectors
v1n = v1 / v1l
v2n = v2 / v2l
print('v1n', np.linalg.norm(v1n))
print('v2n', np.dot(v2n, v2n))
print(np.linalg.norm(v1n), np.linalg.norm(v2n), abs(np.dot(v2n, v1n)))
# Compute the magnitude of the dot product
# Taking abs value as we only want magnitude
d1 = np.abs(np.dot(v1n, v2n))
d2 = np.abs(np.dot(v1, v2))

print(f'dot product unit:', d1)
print('dot product original:', d2)
```
